```python
import numpy as np
```

```python
# q2 - Median, q1 - 1st quartile, q3 - 3rd quartile
q1, q2, q3 = np.percentile(df.charges, [25,50,75])
iqr = q3 - q1  #inter quartikle range
upper_whisker = min(df.charges.max(), q3 +1.5 * iqr) 

lower_whisker = max(df.charges.min(), q1 -1.5 * iqr)

outliers = ((df.charges > upper_whisker) | (df.charges < lower_whisker))
```
