```python
# 1. Find the total number of records
len(df)
```

```python
#2. Find the number of records by stock symbol
df.groupby(df.symbol).symbol.count()
```

```python
# 3. Find avg volume per stock
import numpy as np
df.groupby("symbol").agg({"volume": np.mean})
```

```python
#4. Find Symbol and date that has highest intra day pct chnage in price
import math
def find_intraDay(row):
    return math.fabs(row["close"] - row["open"]) / row["open"]
    
df["IntraDayPCTChange"] = df.apply(find_intraDay, axis = 1)
df.groupby("symbol").agg({"IntraDayPCTChange": np.max}).sort_values("IntraDayPCTChange", ascending = False)
```
